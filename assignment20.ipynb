```python
filepath = 'ford_car_dataset.csv'
df = pd.read_csv(filepath)
```

```python
#independent features
x = df.drop('price', axis=1)
#dependedt features
y = df['price']

print("Independent features :",x.columns)
print("Shape :",x.shape)
print("Dependent features :",y.name)
print("Shape :",y.shape)
```

```python
x_encoded = pd.get_dummies(x, drop_first=False)
x_encoded = x_encoded.astype(int)
print(x_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['year','mileage','tax','mpg','engineSize']

scaler = StandardScaler()
x_encoded[num_cols] = scaler.fit_transform(x_encoded[num_cols])
print(x_encoded[num_cols].head())

```

```python
from sklearn.model_selection import train_test_split

x_train, x_test ,y_train, y_test = train_test_split(x_encoded,
                                             y,
                                             test_size=0.33,
                                             random_state=42)
print("x train shape:",x_train.shape)
print("x test shape:",x_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

print("intercept :",model.intercept_)
print("coefficients :",model.coef_)
```

```python
y_pred = model.predict(x_test)
print("first 10 predict values ")
print(y_pred[:10])

print("first 10 actual values :")
print(y_test.iloc[:10])


comparison = y_test.iloc[:10].reset_index(drop=True).to_frame(name="Actual")
comparison["predicted"] = y_pred[:10]

print("Comparison :")
print(comparison)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R^2 score :",r2)
```

```python
import joblib

joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(x_encoded.columns.tolist(),"columns.pkl")

print("Model, scaler, and columns saved successfully")
```
